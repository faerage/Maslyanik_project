## Контрольні запитання

1. У чому відмінність інтерпретатора від компілятора? Python -- компілятор чи інтерпретатор?

Інтерпретатор виконує програмний код рядок за рядком без попереднього перетворення його в машинний код.
Компілятор перетворює вихідний код програми в машинний код або інший низькорівневий представлення до того, як програма може бути виконана.Python використовує інтерпретатор.

1. Що таке IDE PyCharm?

PyCharm - це інтегроване середовище розробки для мови програмування Python.

1. Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT?

Система контролю версій - це програмне забезпечення, яке призначене для ведення і керування історією змін вихідного коду, документів та інших файлів в проекті. SVN є централізованою системою контролю версій. Це означає, що історія змін та файли зберігаються на центральному сервері, і розробники взаємодіють з цим сервером для витягування останніх змін та внесення своїх власних. GIT є розподіленою системою контролю версій. Кожен розробник має повну копію історії змін на своєму комп'ютері. Це дозволяє робити зміни та відгалужувати проекти навіть офлайн, а потім синхронізувати їх з іншими.

1. Що відбувається при виконанні команди `git add .`?

Команда git add . використовується для додавання всіх змінених та нових файлів в поточному каталозі та всіх його підкаталогах до індексу GIT. Індекс - це проміжна область, де ви готуєте файли для коміту, тобто для фіксації змін у репозиторії GIT.

1. Що відбувається при виконанні команди `git commit`?

Команда git commit використовується для фіксації змін у репозиторії GIT.

1. Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?

Командoю `git push`.
